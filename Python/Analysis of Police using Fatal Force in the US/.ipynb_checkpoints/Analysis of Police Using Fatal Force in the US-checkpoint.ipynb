# Introduction

Since Jan. 1, 2015, [The Washington Post](https://www.washingtonpost.com/) has been compiling a database of every fatal shooting in the US by a police officer in the line of duty. 

<center><img src=https://i.imgur.com/sX3K62b.png></center>

While there are many challenges regarding data collection and reporting, The Washington Post has been tracking more than a dozen details about each killing. This includes the race, age and gender of the deceased, whether the person was armed, and whether the victim was experiencing a mental-health crisis. The Washington Post has gathered this supplemental information from law enforcement websites, local new reports, social media, and by monitoring independent databases such as "Killed by police" and "Fatal Encounters". The Post has also conducted additional reporting in many cases.

There are 4 additional datasets: US census data on poverty rate, high school graduation rate, median household income, and racial demographics. [Source of census data](https://factfinder.census.gov/faces/nav/jsf/pages/community_facts.xhtml).

# Creating a Bar Chart with Subsections Showing the Racial Makeup of Each US State

```python
#removing rows with (X)
df_share_race_city = df_share_race_city[df_share_race_city["share_white"].str.contains("(X)")==False]
```

```python
#convert data types
df_share_race_city[['share_white', 'share_black', 'share_native_american', 'share_asian', 'share_hispanic']] = df_share_race_city[['share_white', 'share_black', 'share_native_american', 'share_asian', 'share_hispanic']].apply(pd.to_numeric)
```

```python
#Grouping by state
df_share_race_city1 = df_share_race_city.groupby('geographic_area', as_index = False).mean()
```

```python
cols = ['share_white', 'share_black', 'share_native_american', 'share_asian', 'share_hispanic']

# getting the total for each state
df_race_sum = df_share_race_city.groupby('geographic_area', as_index=False).sum()
df_race_sum['total'] = df_race_sum.sum(axis=1)

#calculating the sum of each share and then calculating the percentage
df_race_sum.share_white = df_race_sum.share_white / df_race_sum.total * 100
df_race_sum.share_black = df_race_sum.share_black / df_race_sum.total * 100
df_race_sum.share_native_american = df_race_sum.share_native_american / df_race_sum.total * 100
df_race_sum.share_asian = df_race_sum.share_asian / df_race_sum.total * 100
df_race_sum.share_hispanic = df_race_sum.share_hispanic / df_race_sum.total * 100

df_race_sum['total'] = df_race_sum[cols].sum(axis=1)
```

```python
df_race_sum.head()
```

```python
# creating a stack through multiindex
states = []
for state in df_race_sum['geographic_area']:
    count=0
    while count < 5:
        states.append(state)
        count += 1
#creating list of         
share_race = []
for x in range(0, 51):
    for race in cols:
        share_race.append(race)
# using zip for parallel iteration
tuples = list(zip(*[states, share_race]))
index = pd.MultiIndex.from_tuples(tuples, names=['geographic_area', 'Share_Race'])
df = pd.DataFrame(index=index, columns=['pct'])
df.head()
```

```python
# creating number index instead of the MultiIndex categories
df.reset_index(inplace=True)
df.head()
```

```python
df.shape
```

```python
# adding data from df_race_sum into df
race_pct = df_race_sum[cols]
y=0
while y < 255:
    for x in range(0, len(race_pct)):
        for col in race_pct.columns:
                df.pct[y] = race_pct[col][x]
                y += 1

df.head()
```

```python
bar = px.bar(x=df['geographic_area'],
                       y=df.pct,
                       color=df.Share_Race,
                        title='Racial Makeup of Each State')
bar.update_layout(xaxis_title='State',
                             yaxis_title='Percent(%)')
bar.show()
```

Bar graph showing the racial makeup of each US state.

# Were People Armed? 

In what percentage of police killings were people armed? Create chart that show what kind of weapon (if any) the deceased was carrying. How many of the people killed by police were armed with guns versus unarmed? 

```python
df_fatalities_armed = df_fatalities['armed']
df_fatalities_armed = df_fatalities_armed.value_counts().rename_axis('armed').to_frame('counts').reset_index()
```

```python
armed = px.bar(x=df_fatalities_armed.counts,
                y=df_fatalities_armed.armed,
                 orientation='h',
                title='Were People Armed?')
armed.update_layout(xaxis_title='Weapon',
                     yaxis_title='Count')
armed.update_traces(marker_color='black')
armed.show()
```

It is difficult to see the data on this graph because of how high the values for gun, unarmed and knife are. These values are removed for the next graph.

```python
#dropping gun, knife and unarmed to see other weapons better
df_fatalities_armed2 = df_fatalities_armed[df_fatalities_armed['armed'].str.contains('gun')==False]
df_fatalities_armed2 = df_fatalities_armed2[df_fatalities_armed2['armed'].str.contains('knife')==False]
df_fatalities_armed2 = df_fatalities_armed2[df_fatalities_armed2['armed'].str.contains('vehicle')==False]
df_fatalities_armed2 = df_fatalities_armed2[df_fatalities_armed2['armed'].str.contains('unarmed')==False]
df_fatalities_armed2 = df_fatalities_armed2[df_fatalities_armed2['armed'].str.contains('undetermined')==False]
df_fatalities_armed2.head()
```

```python
#new graph
armed = px.bar(x=df_fatalities_armed2.counts,
                y=df_fatalities_armed2.armed,
                
                 orientation='h',
                title='Were People Armed?')
armed.update_layout(xaxis_title='Weapon',
                     yaxis_title='Count')
armed.update_traces(marker_color='black')
armed.show()
```

After the highest values are removed the next main weapon is a toy weapon.

```python
#dropping undetermined
df_fatalities_armed = df_fatalities_armed[df_fatalities_armed.armed != 'undetermined']
df_fatalities_armed = df_fatalities_armed[df_fatalities_armed.armed != 'unknown weapon']
```

```python
#if loop to change values of armed to armed unarmed unknown
for i in df_fatalities_armed['armed']:
    if i == 'toy weapon':
        df_fatalities_armed = df_fatalities_armed.replace(to_replace= i,
                                                         value='unarmed')
    elif i == 'gun':
        df_fatalities_armed = df_fatalities_armed.replace(to_replace = i,
                                                         value = 'armed with gun')
    else:
        df_fatalities_armed = df_fatalities_armed.replace(to_replace = i,
                                                         value = 'armed')
```

```python
#transforming counts to numeric
df_fatalities_armed['counts'] = df_fatalities_armed['counts']. astype(int)
```

```python
df_fatalities_armed3 = df_fatalities_armed.groupby('armed', as_index = False).sum()
df_fatalities_armed3.head()
```

```python
armed = px.bar(x=df_fatalities_armed3.counts,
                y=df_fatalities_armed3.armed,
                
                 orientation='h',
                title='Were People Armed?')
armed.update_layout(xaxis_title='Weapon',
                     yaxis_title='Count')
armed.update_traces(marker_color='black')
armed.show()
```

This graph is showing whether the people being killed by the police were armed, unarmed or armed with gun. Toy weapon has been included in the unarmed catagory, since it is not dangerous and should not have resulted in the death of the person. Armed with gun has been shown seperately as it is the only weapon, in most cases, where shooting is potentially acceptable.

# Number of Police Killings Over Time

Analyse the Number of Police Killings over Time. Is there a trend in the data? 

```python
# removing the day to count just months and years.
df_fatalities.date = pd.to_datetime(df_fatalities.date)
df_fatalities.date = df_fatalities.date.dt.strftime('%Y-%m')
df_fatalities.date = pd.to_datetime(df_fatalities.date)

# counting occurences per month
cases_time = df_fatalities.groupby('date', as_index=False).agg({'name': pd.Series.count})
cases_time.rename(columns={'name': 'cases'}, inplace=True)
cases_time.head()
```

```python
plt.figure(figsize=(14,8), dpi=200)
plt.title('Police Killings Over Time', fontsize=18)
plt.xlabel('Date', fontsize=16)
plt.ylabel('Nr. of Cases', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

ax1 = plt.gca()

ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)

ax1.grid(color='grey', which='both', linestyle='--')

ax1.plot(cases_time.date,
        cases_time.cases,
        linewidth=3)
plt.show()
```

This graph shows a drastic decline in police killings in the middle of 2017.
